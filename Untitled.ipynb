```python
import matplotlib.pyplot as plt
s=['ps','fcsp','de','fsd','ds']
t1=[20,18,22,23,17]
t2=[21,16,22,19,23]
t3=[11,24,13,22,18]
t4=[15,20,17,24,22]
plt.plot(s,t1,'*-y',ms='15',mec='red',mfc='pink',linewidth='4')
plt.plot(s,t2,'*-r',ms='15',mec='red',mfc='pink',linewidth='4')
plt.plot(s,t3,'*-b',ms='15',mec='red',mfc='pink',linewidth='4')
plt.plot(s,t4,'*-g',ms='15',mec='red',mfc='pink',linewidth='4')
a={'family':'serif',
  'color':'red',
  'size':18}
b={'family':'arial',
  'color':'black',
  'size':15}
plt.xlabel('subjects',fontdict=a)
plt.ylabel('Marks',fontdict=b)
plt.title('Data Sheet',fontdict=b,color='blue',loc='left')
plt.grid(axis='y',linestyle='-.',color='black',linewidth='2')
plt.show()
```
