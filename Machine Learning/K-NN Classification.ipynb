```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import numpy as np

# Assuming x and y are your features and labels
rs = range(0, 200)  # Random states range
n_neighbors_range = range(1, 11)  # Number of neighbors range (you can adjust this as needed)

best_accuracy = 0
best_random_state = 0
best_n_neighbors = 0

# Loop over different random states and n_neighbors
for i in rs:
    for n_neighbors in n_neighbors_range:
        # Train-test split with different random states
        x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=i)
        
        # Standardize the data
        s = StandardScaler()
        x_train_scaled = s.fit_transform(x_train)
        x_test_scaled = s.transform(x_test)
        
        # K-Nearest Neighbors classifier
        knn = KNeighborsClassifier(n_neighbors=n_neighbors)
        knn.fit(x_train_scaled, y_train)
        y_pred = knn.predict(x_test_scaled)
        
        # Calculate the accuracy score
        accuracy = accuracy_score(y_test, y_pred)
        
        # Track the best accuracy, random state, and n_neighbors
        if accuracy > best_accuracy:
            best_accuracy = accuracy
            best_random_state = i
            best_n_neighbors = n_neighbors

print(f"Best Accuracy: {best_accuracy}")
print(f"Best Random State: {best_random_state}")
print(f"Best Number of Neighbors: {best_n_neighbors}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors,datasets
from sklearn import neighbors,datasets
from sklearn.preprocessing import StandardScaler
from ipywidgets import interact,fixed

def load_data():
    cancer=datasets.load_breast_cancer()
    return cancer

def plot_boundaries(n_neighbors,data,labels):
    h=0.02
    c_light = ListedColormap(['orange','blue'])
    c_bold = ListedColormap(['darkorange','darkblue'])
    clf=neighbors.KNeighborsClassifier(n_neighbors)
    clf.fit(data,labels)
    
    x_min,x_max = data[:,0].min()-1,data[:,0].max()+1
    y_min,y_max = data[:,1].min()-1,data[:,1].max()+1
    
    xx,yy=np.meshgrid(np.arange(x_min,x_max,h),np.arange(y_min,y_max,h))
    z=clf.predict(np.c_[xx.ravel(),yy.ravel()])
    z=z.reshape(xx.shape)
    
    plt.figure(figsize=(8,6))
    plt.pcolormesh(xx,yy,z,cmap=c_light)
    plt.scatter(data[:,0],data[:,1],c=labels,cmap=c_bold,edgecolor='k',s=20)
    plt.xlim(xx.min(),xx.max())
    plt.ylim(yy.min(),yy.max())
    plt.title(f'2-class classification(k={n_neighbors})')
    plt.show()
cancer=load_data()
x=StandardScaler().fit_transform(cancer.data[:,:2])
y=cancer.target
interact(plot_boundaries,n_neighbors=(1,20),data=fixed(x),labels=fixed(y))
```

```python
#(9-Mark)
df=pd.read_csv("loan_approval_dataset.csv")
print(df.head())
df=df.drop(['loan_id'],axis=1)
df
```

```python
df=df.dropna()
df
```

```python
df=df.drop_duplicates()
df
```

```python
df=df.loc[df['loan_status']!="NA"]
df
```

```python
q1=df['bank_asset_value'].quantile(0.25)
q3=df['bank_asset_value'].quantile(0.75)
iqr=q3-q1
ulimit=q3+1.5*iqr
llimit=q1-1.5*iqr
df=df.loc[(df['bank_asset_value']<=ulimit) & (df['bank_asset_value']>=llimit)]
df
```

```python
import numpy as np
df['education']=np.where(df['education']==' Graduate',1,0)
df['self_employed']=np.where(df['self_employed']==' Yes',1,0)
df
```

```python
x=df.drop(['loan_status'],axis=1)
y=df['loan_status']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
knn=KNeighborsClassifier(n_neighbors=47)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
print("Accuracy score: ",accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import confusion_matrix
a=confusion_matrix(y_test,y_pred)
print(a)
accuracy=((a[0][0]+a[1][1])/(a[0][0]+a[0][1]+a[1][0]+a[1][1]))
print(accuracy)
Errorrate = 1-accuracy
print("Error rate: ",Errorrate)
precision=((a[1][1])/(a[0][1]+a[1][1]))
print("precision: ",precision)
sensitivity=((a[1][1])/(a[1][0]+a[1][1]))
print("sensitivity(recall): ",sensitivity)
specificity=((a[0][0])/(a[0][0]+a[0][1]))
print("specificity: ",specificity)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=20)
from sklearn.tree import DecisionTreeClassifier
decision=DecisionTreeClassifier(criterion='entropy')
decision.fit(x_train,y_train)
y_pred=decision.predict(x_test)
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
