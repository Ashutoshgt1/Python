```python
import socket
mysock=socket.socket()
mysock.connect(('www.data.pr4e.org',80))
cmd='GET http://data.pr4e.org/romeo.txt HTTP/1.0\r\n\r\n'.encode()
mysock.send(cmd)

while True:
  data=mysock.recv(512)
  if(len(data)<1):
    break
  print(data.decode(),end="")
mysock.close()
```

```python
import socket
def createServer():
  ser=socket.socket()
  ser.bind(('localhost',9000))
  ser.listen()
  while True:
    conn,addr=ser.accept()
    rd=conn.recv(5000).decode()

    data="HTTP/1.1 200 OK\n"
    data+="Content-type:text/html;charset=utf-8\n\n"
    data+="<html><body>Hello World</body></html>\n\n"
    conn.send(data.encode())
  ser.close()

print("Access http://localhost:9000")
createServer()
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
response = requests.get("https://www.politifact.com/factchecks/list/")
content = response.text
soup = BeautifulSoup(content,'html.parser')
#print(soup.prettify())

source=soup.find_all('a',class_="m-statement__name")
statement=soup.find_all('div',class_="m-statement__quote")
date=soup.find_all('footer',class_="m-statement__footer")
img=soup.find_all('img',class_="c-image__original",alt=True)

d={'source':[],'statement':[],'date':[],'img':[]}
for i in range(len(source)):
  d['source'].append(source[i].text.strip())
  d['statement'].append(statement[i].text.strip())
  d['date'].append(date[i].text.split("•")[1].strip)
  d['img'].append(img[i]['alt'])

df=pd.DataFrame(d)
df

```
