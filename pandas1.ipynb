```python
d=[{'a':1,'b':2,'c':3,'d':4},
   {'a':10,'b':20,'c':30,'d':40},
   {'a':100,'b':200,'c':300,'d':400},]
df=pd.DataFrame(d)
print(df.iloc[0:1])           # slicing do not rdeuce 2d to 1d
print(type(df.iloc[0:1]))    
print(df.iloc[0,1])  #indexing does reduce the 2d to 1d to element
print(df.iloc[[0],[1,2]])
print(df.iloc[[True,False,True]])
print(df.iloc[lambda x:x.index%2==0])
df
```

```python
df=pd.DataFrame([[1,2],[4,5],[7,8]],index=['A','B','C'],columns=['X','Y'])
print(df.loc['B'])    #indexing ma series
print()
print(df.loc['B':'C'])#slicing ma dataframe
print()
print(df.loc[['A','C'],'Y'])
print()
print(df.loc[df['Y']>5])# returns whole row in which y>5
print(type(df.loc[df['Y']>5]))
print()
print(df.loc[df['Y']>5,'X'])
print()
print(df.loc[(df['Y']>5)|(df['X']<5)])
print()
print(df.loc[(df['Y']>5)&(df['X']<5)])
print()
print(df.loc[df['Y']>6,['X']])#list pass karyu chhe which works as slicing so no reducing
print()
print(df.loc[lambda df:df['Y']==8])
print()
df.loc[['B','C'],['Y']]=50
df
```

```python
#if there is null value in a or c column then it drops that row
df.dropna(subset=['A','C'])
```

```python
#drop rows which have <2 non-nan values
df.dropna(thresh=2)
```
