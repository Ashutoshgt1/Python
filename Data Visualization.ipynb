# Area Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
data={'year':[2015,2016,2017,2018,2019,2020],
     'sales':[100,60,150,170,200,210],
     'Expenses':[70,80,90,100,110,120]}
df=pd.DataFrame(data)
plt.stackplot(df['year'],df['sales'],df['Expenses'],labels=['sales','Expenses'])
plt.legend(loc='upper left')
plt.grid()
plt.show()
plt.fill_between(df['year'],df['sales'],df['Expenses'])
plt.show()
```

```python
months=np.arange(1,13)
sales=np.array([10,15,20,30,20,50,30,70,80,40,100,10])
expenses=np.array([5,10,15,20,25,30,35,40,45,50,55,60])
plt.plot(months,sales,color='green',label='sales')
plt.plot(months,expenses,color='red',label='expenses')
plt.fill_between(months,sales,expenses,where=sales>expenses,alpha=.25,color='green',interpolate=True)
plt.fill_between(months,sales,expenses,where=sales<expenses,alpha=.25,color='red',interpolate=True)
plt.legend()
plt.savefig("myimage.png",format='png')#write before show()
plt.show()
```
